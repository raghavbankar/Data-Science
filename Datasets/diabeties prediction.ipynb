## Diabeties Prediction System

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
```

```python
Diabetes=load_diabetes()
```

```python
Diabetes
```

```python
print(Diabetes['DESCR'])
```

```python
## Independent features
X=pd.DataFrame(Diabetes['data'],columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5',"s6"])
```

```python
#Dependent features
Y=Diabetes['target']
```

```python
Y
```

```python
## Train Test Split

from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=10,test_size=0.2)
```

```python
## making the model
from sklearn.tree import DecisionTreeRegressor
regression=DecisionTreeRegressor()
```

```python
## Hyperparameter Tunning

params={
    'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter':['best', 'random'],
    'max_depth':[2,5,6,7,8,9,10],
    'max_features':['sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
Grid=GridSearchCV(regression,param_grid=params,cv=5,refit=True,scoring='neg_mean_absolute_error')
```

```python
Grid.fit(X_train,Y_train)
```

```python
Grid.best_params_
```

```python
Y_pred=Grid.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(mean_absolute_error(Y_test,Y_pred))
print(mean_squared_error(Y_test,Y_pred))
print(r2_score(Y_test,Y_pred))
```
