```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
iris
```

```python
##Independent Features
X=pd.DataFrame(iris['data'],columns=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'])
```

```python
##Dependent Features
Y=pd.DataFrame(iris['target'])
Y.value_counts()
```

Testing data is used to determine the performance of the trained model, whereas training data is used to train the machine learning model. 

```python
##train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

## Decision Tree

A decision tree consists of a set of decision nodes, connected by branches,
extending downward from the root node until terminating in leaf nodes.



Entropy can be defined as a measure of the purity of the sub-split. Entropy always lies between 0 to 1.

It is the probability of misclassifying a randomly chosen element in a set.
used for building the tree by splitting as per the appropriate features

 Information gain is a measure of this change in entropy.

 Greatest IG feature will be selected for the root node

#No normalization required in decision tree

## Cart Decision Tree

```python
##Applying Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
treeclassifier=DecisionTreeClassifier(criterion='gini')
```

```python
Y_train
```

```python
treeclassifier.fit(X_train,Y_train)
```

```python
##Visualizing the Decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
y_pred=treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(Y_test,y_pred))
print(confusion_matrix(Y_test,y_pred))
print(classification_report(Y_test,y_pred))
```
